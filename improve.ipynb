```
!git clone https://github.com/telecombcn-dl/2018-dlai-team4.git
import os
os.chdir("2018-dlai-team4")

from import_picture import import_processed_pict_from

size_of_output=100                #square image so size_of_output*size_of_output
#nomenclature : "filtered_pict_" + how to choose the people (number of picture each people should have at least) + number of pict kept for each person (at maximum)
#folder_to_use="filtered_pict_20"
#folder_to_use="filtered_pict_10_every-picture" 
folder_to_use="filtered_pict_10_10" 
RGB_output=True
run_preprocessing = True


(x_train, y_train), (x_val, y_val), (x_test, y_test),name_list=import_processed_pict_from(folder_to_use,size_of_output,RGB_output)
```

```
print("Totaly picture counts : "+str(len(x_train)+len(x_val)+len(x_test)))
print("Output Label Sample : "+str(y_train))
```

```
import numpy as np
from keras.utils import to_categorical

# array transformation
x_train=np.asarray(x_train)
x_test=np.asarray(x_test)
x_val=np.asarray(x_val)


# image resize ( turn to gray scale)
if not RGB_output: 
  x_train=x_train.reshape(len(x_train),size_of_output,size_of_output,1)
  x_test=x_test.reshape(len(x_test),size_of_output,size_of_output,1)
  x_val=x_val.reshape(len(x_val),size_of_output,size_of_output,1)

# Categorization process
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)

```

Sample picture representation ( just take into account face of the person by using special library)

```
ind = np.arange(len(x_train))   # creating index number as much as length of x_train
np.random.shuffle(ind)          # mixing data

print("Before shuffle: "+str(y_train[1]))

x_train = x_train[ind]
y_train = y_train[ind]

print("\n\nAfter shuffle: "+str(y_train[0]))


print("\n\nClassification counts :"+ str(len(y_train[0])))
```

# **Same CNN with data augmentation**

```
from keras import layers
from keras import models
from keras import regularizers

#new design approaching  

# Creating Network

output_size=len(y_train[1])
print("output size automatisation : "+ str(output_size))

input_size=x_train[1].shape
print("input size automatisation : "+ str(input_size))

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',input_shape=input_size))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))                                 # To avoid overfitting
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(output_size, activation='softmax'))     # Look at activation function

```

```
model.summary()
```

```
from keras import optimizers

#opt = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)  #optimizer choosing
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

## Data Augmentantion

Apply preprocessing on training data

```
from keras.preprocessing.image import ImageDataGenerator , array_to_img, img_to_array, load_img
from matplotlib import pyplot
from keras.preprocessing import image




# New approaching

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')


train_datagen.fit(x_train, augment=True)

i=0
for batch in train_datagen.flow(x_train, batch_size=5):
  
    
    #plt.imshow(x_batch[0])
    
    #print(batch[0])
    #imgplot = plt.imshow(x_train[0].array_to_img(batch[0]))
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[i]))
    i += 1
    if i % 4 == 0:
        break

plt.show()


```

Train the model

```
epoc = 100

history = model.fit_generator(train_datagen.flow(x_train, y_train,batch_size = 32,shuffle = True),  
                    epochs=epoc,
                    shuffle = True, 
                    steps_per_epoch =  len(x_train)/32,
                    validation_data = (x_val,y_val),
                    validation_steps = 50)
```

```
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)


# "bo" is for "blue dot"
plt.plot(epochs, acc, 'bo', label='Training accuracy')
# b is for "solid blue line"
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

Compute the test accuracy

```
test_loss, test_acc = model.evaluate(x_test, y_test)

print("Test Data Count : " +str(len(x_test)))
print("Accuracy of Test Data : "+ str(test_acc))
```
